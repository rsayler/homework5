## Total Fares by City Type

```python
# Calculate Type Percents

# Build Pie Chart
city_types = ["Urban", "Rural", "Suburban"]
city_percentages = [62.7, 6.8, 30.5]
colors = ["lightcoral", "gold", "lightblue"]
explode = (0.1,0,0)
plt.title("% of Total Fares by City Type")
plt.pie(city_percentages, explode=explode, labels=city_types, colors = ["lightcoral", "gold", "lightblue"], autopct="%1.1f%%", shadow=True, startangle=270)

plt.axis("equal")

# Save Figure
plt.savefig("../Images/Fares_by_city")
            
```
